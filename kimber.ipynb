```python
#import files
import pandas as pd
```

```python
df1 = pd.read_csv(r'C:\Users\shaya\Downloads\t20i_deathbowl.csv')
print(df1)
```

```python
df2 = pd.read_csv(r'C:\Users\shaya\Downloads\t20_deathbowl.csv')
print(df2)
```

```python
#define df3 for merging common players files 
df3 = pd.merge(df1, df2[['Player', 'O', 'R', 'W']], on='Player')
```

```python
print(df3)
```

```python
df3['O'] = df3['O_x'] + df3['O_y']
df3['R'] = df3['R_x'] + df3['R_y']
df3['W'] = df3['W_x'] + df3['W_y']

print (df3)
```

```python
df3 = df3[['Player','O','R','W']]
df3
```

```python
df3['Avg']=df3['R']/df3['W']
df3['B']= df3['O']%1*10 + (df3['O'] - (df3['O']%1))*6
df3['Econ']=df3['R']*6/df3['B']
df3['SR']=df3['B']/df3['W']
df3
```

```python
df_combined = pd.concat([df1, df2], ignore_index=True)
df_unique = df_combined.drop_duplicates(subset='Player', keep=False)
df4 = df_unique
df4['Avg']=df4['R']/df4['W']
df4['B']= df4['O']%1*10 + (df4['O'] - (df4['O']%1))*6
df4['Econ']=df4['R']*6/df4['B']
df4['SR']=df4['B']/df4['W']
df4
```

```python
df5 = pd.concat([df3, df4], axis=0)
print(df5)
```

```python
df5.to_csv('deathbowl_overall.csv', index=True)
```
